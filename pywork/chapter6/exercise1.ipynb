## 연습문제 1번

1. 0~999의 세 자리 숫자로 이루어진 다음 리스트를 기수 정렬을 이용해 정렬하는 과정을 설명하라.  
   
        [10, 123, 56, 636, 992, 119, 234, 76, 82, 345, 567, 432] 

기수 정렬(Radix Sort): 낮은 자리수부터 높은 자리수로 정렬을 수행하는 방식

1. 일의 자리수부터 시작하여 각 자리수를 기준으로 정렬
2. 십의 자리수를 기준으로 정렬
3. 백의 자리수를 기준으로 정렬

### Step 1: 일의 자리수 정렬

- 일의 자리수가 0인 수: 10
- 일의 자리수가 2인 수: 82, 432
- 일의 자리수가 3인 수: 123
- 일의 자리수가 4인 수: 234
- 일의 자리수가 5인 수: 56, 345
- 일의 자리수가 6인 수: 636, 76
- 일의 자리수가 7인 수: 567
- 일의 자리수가 9인 수: 119, 992

정렬된 리스트: [10, 432, 82, 123, 234, 345, 56, 636, 76, 567, 119, 992]

### Step 2: 십의 자리수 정렬

- 십의 자리수가 0인 수: 10
- 십의 자리수가 2인 수: 123, 234
- 십의 자리수가 3인 수: 432, 345
- 십의 자리수가 5인 수: 56
- 십의 자리수가 6인 수: 636, 567
- 십의 자리수가 7인 수: 76
- 십의 자리수가 8인 수: 82
- 십의 자리수가 9인 수: 992, 119

정렬된 리스트: [10, 123, 234, 432, 345, 56, 636, 567, 76, 82, 992, 119]

### Step 3: 백의 자리수 정렬

- 백의 자리수가 0인 수: 10, 56, 76, 82
- 백의 자리수가 1인 수: 119, 123
- 백의 자리수가 2인 수: 234
- 백의 자리수가 3인 수: 345
- 백의 자리수가 4인 수: 432
- 백의 자리수가 5인 수: 567
- 백의 자리수가 6인 수: 636
- 백의 자리수가 9인 수: 992

정렬된 리스트: [10, 56, 76, 82, 119, 123, 234, 345, 432, 567, 636, 992]

기수 정렬이 끝난 리스트: [10, 56, 76, 82, 119, 123, 234, 345, 432, 567, 636, 992]
