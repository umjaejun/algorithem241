2. 출력

## 5. 코드

```python
def radix_sort(A) :
    
    queues = []
    for i in range(BUCKETS) :
        queues.append(Queue())

        n = len(A)
        factor = 1
        for d in range(DIGITS) :
            for i in range(n) :
                queues[(A[i]//factor) % BUCKETS].put(A[i])
            j = 0
            for b in range(BUCKETS) :
                while not queues[b].empty() :
                    A[j] = queues[b].get()
                    j += 1
            factor *= BUCKETS
            print("step", d+1, A)
```

## 6. 테스트 코드

```python
from queue import Queue

def radix_sort(A):
    BUCKETS = 10
    DIGITS = len(str(max(A)))
    
    queues = []
    for i in range(BUCKETS):
        queues.append(Queue())

    n = len(A)
    factor = 1
    for d in range(DIGITS):
        for i in range(n):
            queues[(A[i] // factor) % BUCKETS].put(A[i])
        j = 0
        for b in range(BUCKETS):
            while not queues[b].empty():
                A[j] = queues[b].get()
                j += 1
        factor *= BUCKETS
        print("step", d+1, A)

A = [170, 45, 75, 90, 802, 24, 2, 66]

radix_sort(A)
```

## 8. 복잡도 분석

시간 복잡도는 O(d * (n + k))
여기서 d는 숫자의 최대 자릿수, n은 배열의 크기, k는 숫자 범위를 나타낸다.

시간 복잡도 분석

- d (최대 자릿수): Radix Sort는 숫자의 각 자릿수를 기준으로 정렬을 수행합니다. 따라서 최대 자릿수 d만큼 정렬 과정을 반복해야 한다.
- n (배열 크기): 각 자릿수 정렬 시 n개의 원소를 처리해야 한다.
- k (숫자 범위): 각 자릿수 정렬 시 0부터 k-1까지의 숫자 범위에 대해 계수 정렬(Counting Sort)을 수행합니다. 계수 정렬의 시간 복잡도는 O(n+k)이다.
따라서 Radix Sort의 전체 시간 복잡도는 d번의 계수 정렬을 수행하므로 O(d * (n + k))가 된다.
