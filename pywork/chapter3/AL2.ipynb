## 문제 2번

## 1. 문제 정의
리스트 선택 정렬을 이용한 오름차순 정렬

## 2. 알고리즘 설명

위 알고리즘 은 선택 정렬 알고리즘으로
 - 리스트에서 첫 번째 요소를 선택
 - 나머지 요소들 중에서 최소값(또는 최대값)을 찾는다
 - 최소값(또는 최대값)을 선택한 요소와 교환
 - 다음 위치에 있는 요소를 선택하고, 위의 과정을 반복
 - 위의 과정을 리스트의 끝까지 반복하여 정렬이 완료

이 알고리즘은 각 반복에서 리스트의 나머지 부분 중 최소값(또는 최대값)을 찾기 위해 선형 탐색을 수행 이러한 과정을 반복하며 전체 리스트를 정렬함

## 3. 손으로 푼 예제

## 4. 알고리즘 개요

입력 변수
 - arr

이 알고리즘은 arr 을 직접 수정하여 정렬된 상태로 만들기 때문에 별도의 출력 변수는 존재하지 않는다.

## 5. 알고리즘 코드

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
```

## 6. 테스트 코드

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

arr = [7, 4, 9, 6, 3, 8, 7, 5]
selection_sort(arr)
print("정렬된 리스트:", arr)
```

## 8. 복잡도 분석

위 알고리즘의 복잡도는 O(n^2)이다.

이 알고리즘은 리스트의 길이에 비례하여 제곱 비례로 시간 복잡도가 증가한다.

2개의 반복문을 사용하여 구현되며, 바깥쪽 반복문은 n번 반복, 안쪽 반복문은 현재 위치릐 요소와 나머지 요소를 비교하며 최솟값을 찾기 위해 n-1, n-2, .... 1 번 반복한다.

따라서 비교 횟수는 (n-1) + (n-2) + .... + n(n-1)/2 이다.
