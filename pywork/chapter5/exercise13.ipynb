## 연습문제 13번

13. 다음의 정수 배열을 퀵 정렬을 이용해 오름차순으로 정렬하려고 한다. median of three를 피벗으로 사용하는 경우 각 단계에서의 배열의 내용을 적어라.  

                    1   2   3   4   5   6   7   8               

## 첫 번째 단계
- l = 0, r = 7
- m = (l + r) // 2 = 3 에 의해, 피벗: 4
- 피벗 4를 오른쪽 끝으로 이동: [1, 2, 3, 8, 5, 6, 7, 4]
- store_index = 0부터 피벗 4보다 작은 값들을 왼쪽으로 이동: [1, 2, 3, 4, 5, 6, 7, 8]
- 피벗을 제 위치에 놓기: [1, 2, 3, 4, 5, 6, 7, 8]
- 분할 결과: 좌측 [1, 2, 3], 우측 [5, 6, 7, 8]

## 두 번째 단계
- 좌측 부분 배열 분할
- l = 0, r = 2
- m = (l + r) // 2 = 1, 피벗: 2
- 피벗 2를 오른쪽 끝으로 이동: [1, 3, 2, 4, 5, 6, 7, 8]
- store_index = 0부터 피벗 2보다 작은 값들을 왼쪽으로 이동: [1, 2, 3, 4, 5, 6, 7, 8]
- 피벗을 제 위치에 놓기: [1, 2, 3, 4, 5, 6, 7, 8]
- 분할 결과: 좌측 [1], 우측 [3]

## 세 번째 단계
- 우측 부분 배열 분할
- l = 4, r = 7
- m = (l + r) // 2 = 5, 피벗: 6
- 피벗 6를 오른쪽 끝으로 이동: [1, 2, 3, 4, 5, 8, 7, 6]
- store_index = 4부터 피벗 6보다 작은 값들을 왼쪽으로 이동: [1, 2, 3, 4, 5, 6, 7, 8]
- 피벗을 제 위치에 놓기: [1, 2, 3, 4, 5, 6, 7, 8]
- 분할 결과: 좌측 [5], 우측 [7, 8]

# 네 번째 단계
- 우측 부분 배열 [7, 8] 분할
- l = 6, r = 7
- m = (l + r) // 2 = 6, 피벗: 7
- 피벗 7를 오른쪽 끝으로 이동: [1, 2, 3, 4, 5, 6, 8, 7]
- store_index = 6부터 피벗 7보다 작은 값들을 왼쪽으로 이동: [1, 2, 3, 4, 5, 6, 7, 8]
- 피벗을 제 위치에 놓기: [1, 2, 3, 4, 5, 6, 7, 8]
- 분할 결과: 좌측 [7], 우측 [8]
