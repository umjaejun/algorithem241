## 2. 알고리즘 설명
- 세 요소의 중간 값을 선택하여 해당 요소의 인덱스를 반환하는 알고리즘.
- l은 배열의 왼쪽 끝, r은 배열의 오른쪽 끝 인덱스를 나타낸다.
  
1. 중간 인덱스 계산
   - m = (l+r) // 2
2. 중간 값 판단
   - if ((a[l] < a[m] and a[m] < a[r]) or (a[r] < a[m] and a[m] < a[l])): return m
   - a[l], a[m], a[r] 중에서 a[m]이 중간 값인지 확인한다. a[l] < a[m] < a[r] 또는 a[r] < a[m] < a[l]인 경우 m이 중간 값이 된다.
3. 왼쪽 값 판단
   - if ((a[m] < a[l] and a[l] < a[r]) or (a[r] < a[l] and a[l] < a[m])): return l
   - a[l]이 중간 값인지 확인한다. a[m] < a[l] < a[r] 또는 a[r] < a[l] < a[m]인 경우 l이 중간 값이 된다.
4. 오른 쪽 값 반환
   - 두 조건에 해당하지 않으면 a[r]이 중간 값이므로 r을 반환한다.

## 5. 코드

```python
def median_of_three(a, l, r) :
    m = (l+r) // 2
    if ((a[l] < a[m] and a[m] < a[r]) or (a[r] < a[m] and a[m] < a[l])) :
        return m
    if ((a[m] < a[l] and a[l] < a[r]) or (a[r] < a[l] and a[l] < a[m])) :
        return l
    else :
        return r
```

## 6. 테스트 코드

```python
def median_of_three(a, l, r):
    m = (l + r) // 2
    if ((a[l] < a[m] and a[m] < a[r]) or (a[r] < a[m] and a[m] < a[l])):
        return m
    if ((a[m] < a[l] and a[l] < a[r]) or (a[r] < a[l] and a[l] < a[m])):
        return l
    else:
        return r

#4.12의 partition()
def quicksort(a, l, r):
    if l < r:
        # 피벗 선택
        pivot_index = median_of_three(a, l, r)
        pivot_value = a[pivot_index]
        # 피벗을 오른쪽 끝으로 이동
        a[pivot_index], a[r] = a[r], a[pivot_index]
        store_index = l
        # 피벗보다 작은 요소들은 왼쪽으로 이동
        for i in range(l, r):
            if a[i] < pivot_value:
                a[i], a[store_index] = a[store_index], a[i]
                store_index += 1
        # 피벗을 제 위치에 놓기
        a[store_index], a[r] = a[r], a[store_index]
        # 재귀적으로 좌우 부분 배열 정렬
        quicksort(a, l, store_index - 1)
        quicksort(a, store_index + 1, r)

# 배열 정렬
array = [1, 2, 3, 4, 5, 6, 7, 8]
quicksort(array, 0, len(array) - 1)
print(array)
```

   - if ((a[m] < a[l] and a[l] < a[r]) or (a[r] < a[l] and a[l] < a[m])):
   - 두 개의 if 조건문과 하나의 else 블록으로 구성된다. 각 조건문 안에는 두 개의 비교 연산과 하나의 논리 연산이 포함되어 있다.
   - 각 비교 연산과 논리 연산은 상수 시간에 이루어진다. 따라서 복잡도는 O(1)
3. 결론
   - 각 단계가 상수 시간(O(1))에 실행되므로, 전체 함수의 시간 복잡도는 O(1)
