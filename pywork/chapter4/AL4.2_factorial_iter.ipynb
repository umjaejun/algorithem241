## 예제 4.2

## 1. 문제 정의
팩토리얼(상향식 축소 정복 기법)

## 2. 알고리즘 설명
- 반복문을 사용하여 주어진 숫자의 팩토리얼 계산하는 알고리즘.
- 1 ~ n 까지의 모든 자연수를 곱한 값을 계산한다.

factorial_iter 함구로 구현되어 있다.

factorial_iter는 한대의 매개 변수 n을 받고 , result라는 변수 초기값 1로 설정한다.

for 문을 사용하여 1 ~ n 까지의 모든 값을 반복하며, result에 현재의 수를 곱하여 갱신한다. 

result에 1 * 2 * ... * n 의 결과가 저장된다.

result 값을 반환하며, 결과를 반환한다.

## 3. 손으로 푼 예제

## 4. 알고리즘 개요
- 입력 변수(input)

  n : 팩토리얼 계산할 자연수
  
- 출력(Output)

  result : 입력된 자연수 n 의 팩토리얼 값
  

## 5. 알고리즘 코드

```python
def factorial_iter(n) :
    result = 1
    for k in range(1, n+1) :
        result = result * k
    return result
```

## 6. 테스트 코드

```python
def factorial_iter(n) :
    result = 1
    for k in range(1, n+1) :
        result = result * k
    return result

print(factorial_iter(10))
```

## 8. 복잡도 분석
for문 사용되며 1 ~ n 까지의 숫자를 반복한다. 따라서 시간 복잡도는 O(n)이다.

즉 n에 비례하여 계산 시간이 증가한다.
